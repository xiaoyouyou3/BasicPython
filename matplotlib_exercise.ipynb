# 【確認課題】Matplotlib

```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
hist_data = pd.read_csv("./data/hist_data.csv")
plt.figure(figsize=(4, 3))
plt.hist(hist_data["aデータ"], bins=30, alpha=0.5, label="aデータ", color="orange")
plt.hist(hist_data["bデータ"], bins=30, alpha=0.5, label="bデータ", color="#9fefe7")
plt.xticks(fontsize=7)
plt.xlabel("温度(℃)")
plt.xticks(fontsize=7)
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布", fontsize="small")
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
scatter_data = pd.read_csv("./data/scatter_data.csv")
scatter_data["class"] = scatter_data["class"].replace({0:"class0", 1:"class1", 2:"class2"})
colors = ["red", "blue", "green"]

plt.figure(figsize=(4,3))
sns.scatterplot(data=scatter_data, x="length (cm)", y="width (cm)", hue="class", palette=colors)
plt.title("クラス0～2のlengthとwidthに関する散布図", fontsize="small")
plt.legend(loc="upper right", fontsize=7, markerscale=0.7)
plt.xticks(np.arange(4.5, 8.5, 0.5), fontsize=7)
plt.xlabel("length (cm)", fontsize=7)
plt.yticks(fontsize=7)
plt.ylabel("width (cm)", fontsize=7)

plt.grid()
plt.subplots_adjust(left=0.01, bottom=0.01)

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

```python
line_data = pd.read_csv("./data/line_data.csv")

x = line_data["期間"]
yplot = line_data["件数"]

plt.figure(figsize=(4,3))
plt.plot(x, yplot)
plt.scatter(x, yplot, s=14)
plt.title("期間ごとの件数の推移", fontsize=7)
plt.xticks(rotation=30, fontsize=7)
plt.xlabel("期間", fontsize=7)
plt.yticks(np.arange(105, 145, 5), fontsize=7)
plt.ylabel("件数", fontsize=7)
plt.grid()
plt.show()
```

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
ybar = line_data["費用"]

plt.figure(figsize=(4,3))
plt.bar(x, ybar, align="center", width=0.8)
plt.title("期間ごとの費用の水位", fontsize=7)
plt.xticks(rotation=30, fontsize=7)
plt.xlabel("期間", fontsize=7)
plt.yticks(fontsize=7)
plt.ylabel("費用(円)", fontsize=7)
plt.grid()
plt.show()
```
